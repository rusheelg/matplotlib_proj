```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Visualization for Year-wise Gas Prices Data**

```python
gas_data = "https://raw.githubusercontent.com/rusheelg/matplotlib_proj/main/gas_prices.csv"
gas = pd.read_csv(gas_data)

plt.figure(figsize=(12,6), dpi=80)
plt.title("Gas prices over time in USD", fontdict={'fontweight':'bold','fontsize':12})
plt.plot(gas.Year, gas.USA, label="United States", linestyle="--",marker=".")
plt.plot(gas.Year, gas.Canada, label="Canada", color = "red" , marker=".")
plt.plot(gas.Year, gas['South Korea'],color="green",marker=".", label="South Korea")
plt.plot(gas.Year, gas.Australia, label="Australia", color = "yellow" , marker=".")
plt.xticks(gas.Year[::3].tolist() + [2011])
plt.xlabel("Year",fontdict={'fontweight':'bold'})
plt.ylabel("US Dollars",fontdict={'fontweight':'bold'})
plt.legend()
plt.show()

plt.savefig("gas price graph.png", dpi=80)
```

**Visualiation for FIFA-18 Player data**

```python
#fifa 18 player data
fifa_raw = "https://raw.githubusercontent.com/rusheelg/matplotlib_proj/main/fifa_data.csv"
fifa = pd.read_csv(fifa_raw)
```

```python
#Histogram Plot for Distribution for Overall Player skills in FIFA 2018
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins,color="#abcdef")
plt.xticks(bins)
plt.ylabel("No of players")
plt.xlabel("Skill level")
plt.title("Distribution of player skills in FIFA 2018")
plt.show()
```

```python
#Pie chart for Foot Preferences of Players
plt.figure(figsize=(8,8))
left = fifa.loc[fifa["Preferred Foot"] == 'Left'].count().Name
right = fifa.loc[fifa["Preferred Foot"] == 'Right'].count().Name
plt.pie([left,right], labels=['Left','Right'], colors=["#66ff66","#33ccff"],autopct='%.1f%%')
plt.title("Foot preference of FIFA players")
plt.show()
```

```python
# Pie chart for weight distribution of players

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]
weights=[light,light_medium,medium,medium_heavy,heavy]

plt.figure(figsize=(10,10),dpi=80)
plt.style.use('ggplot')
explode=(.4,.2,0,0,.4)
labels=['Under 125','125 - 150','150 - 175','175 - 200','Greater than 200']
plt.pie(weights, labels=labels, autopct='%.2f%%',pctdistance=0.8,explode=explode)
plt.title("Weight Distribution of FIFA players in lbs")
plt.show()

```

```python
# Boxplot for Overall rating of Soccer Teams
barca = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
real_madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
man_city = fifa.loc[fifa.Club == 'Manchester City']['Overall']
club=[barca,real_madrid,man_city]

labels=['FC Barcelona','Real Madrid','Manchester City']
plt.figure(figsize=(5,9))
plt.style.use('default')
boxes = plt.boxplot(club,labels=labels)
#setting colors of the boxes
for box in boxes['boxes']:
  box.set(color='blue', linewidth = 2)

plt.title("Professional Soccer team comparision")
plt.ylabel("Overall Rating")
plt.show()
```
